### Chi-square test to find the Relationship between Gender and Specialization
- Is there a relationship between gender and specialization? (i.e., Does the preference of Specialization depend on the Gender?)
- Lets make a bold Claim that Specialization depend on the Gender.

Step - 1:
* Null Hypothesis(H0): Specialization independent on the Gender
* Alternate Hypothesis(H1): Specialization dependent on the Gender


```python
from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
df = data[['Gender','Specialization']]
df.head()
```

```python
data_crosstab = pd.crosstab(index=df.Specialization, columns=df.Gender)
data_crosstab
```

```python
chi2_contingency(data_crosstab)
```

```python
chi2_test_stat = chi2_contingency(data_crosstab)[0]
pval = chi2_contingency(data_crosstab)[1]
df = chi2_contingency(data_crosstab)[2]
```

```python
alpha = 0.05

chi2_critical = chi2.ppf(1 - alpha, df)

chi2_critical
```

```python
if(chi2_test_stat > chi2_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
if(pval < alpha):
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

Conclusion: 
* After Performing the Chi_square test on the Specialization and Gender we can say that these two features are dependent on each other.
